```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Extracting the data and combining it into the dataFrame
```

```python
df_train=pd.read_excel('Data_Train.xlsx')
df_test=pd.read_excel('Test_set.xlsx')
df=pd.concat([df_train,df_test],ignore_index=True)
```

```python
df.head()
```

```python
df.info()
```

```python
#Extracting the Date of Journey into date,month,year
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y', errors='coerce', dayfirst=True)
```

```python

df['Date']=df['Date_of_Journey'].dt.day
df['Month'] = df['Date_of_Journey'].dt.month
df['Year'] = df['Date_of_Journey'].dt.year
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
#Extracting the Arrival_Time into Hour,Minutes
```

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minutes']=df['Arrival_Time'].str.split(':').str[1]
```

```python
#changing the datatype
```

```python
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
```

```python
df['Arrival_Minutes']=df['Arrival_Minutes'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Extracting the Dep_Time into Hour,Minutes
```

```python
df['Dep_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Minutes']=df['Dep_Time'].str.split(':').str[1]
```

```python
#changing the datatype into int
```

```python
df['Dep_Hour']=df['Dep_Hour'].astype(int)
df['Dep_Minutes']=df['Dep_Minutes'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```
